```python
## Question 4(a) approx. value using Recursion:
def a1q4(n):
  if n == 0:
    return 1
  if n == 1:
    return 0.2857142857
  else:
    return (18*a1q4(n-2) - 49*a1q4(n-1)) / 49

# tests:
a1q4(0),a1q4(1),a1q4(2),a1q4(3),a1q4(5)
```

```python
## Question 4(a) approx value using Loop:
def a1(n):
  n = n+1
  L = list(range(n))
  for i in range(n):
    if i < 2:
      L[0] = 1
      L[1] = 0.2857142857
    elif i >= 2:
      L[i] = (18*L[i-2] - 49*L[i-1]) / 49
  return L[i]

# tests:
a1q4(0),a1q4(1),a1q4(2),a1q4(3),a1q4(5)
```

```python
## calculate the absolute value of the exact value and the approx.
##   using the recursion function a1q4(n):
## va is the approximated value; ve is the exact value
def ab_error(va,ve):
  ab = abs(va-ve)
  return ab/ve

# test:
n=17
va = a1q4(n)
ve = (2/7)**n
va, ve, abs(va-ve), ab_error(va,ve)
```

```python
## plot of the absolute error from n=0 to n=20
import matplotlib.pyplot as plt
x = [ab_error(a1q4(0),(2/7)**0),ab_error(a1q4(5),(2/7)**5),ab_error(a1q4(10),(2/7)**10),
     ab_error(a1q4(20),(2/7)**20)]
y = [0,0.01,0.02,0.03]
plt.plot(x,y)
plt.xlabel('absolute error')
plt.title('Absolute Error as n increases')
plt.show()
```
